```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the data
df = pd.read_csv("EATS Analytics Exercise - Q1 Data - Sheet1.csv - EATS Analytics Exercise - Q1 Data - Sheet1.csv", index_col="Courier Payments")
del df["Unnamed: 1"] # there where a blank column in the data
df
```

```python
courierPayments = df.transpose()
courierPayments.reset_index(inplace=True)
courierPayments.rename(columns = {"index": "Date"}, inplace=True)
courierPayments
```

```python
#we brought the data set to a more understandable model
# lets see whats on it 
courierPayments.info()
```

```python
#this cols just change its type["Deliveries", "Gross Booking", "Restaurant Payments", "Courier Payments", "Eater Fre Adjusments", "Eater Referral Promotions", "Delivery Fee"]
cols = ["Deliveries", "Gross Bookings", "Restaurant Payments", "Courier Payments", "Eater Fare Adjustments","Eater Other Promotionss / Discounts", "Eater Referral Promotions", "Delivery Fee"]
def fixCurrencCols(df, cols):
    for col in cols:
        df[col] = df[col].str.replace(",", "")
        df[col] = df[col].str.replace("$", "")
        df[col] = df[col].astype(int)
    return df

courierPayments_fix = fixCurrencCols(courierPayments, cols)
courierPayments_fix["Date"] = pd.to_datetime(courierPayments_fix["Date"])
display(courierPayments_fix)
display(courierPayments_fix.info())
#The data is in proper format to be analyzed
```

```python
#What are the delivery unit economics of this city (e.g. how much money do we make or lose on
#average for each order) and how do they evolve over time?
courierPayments_fix["Unit Economics"] = (courierPayments_fix["Gross Bookings"] - courierPayments_fix["Restaurant Payments"] - courierPayments_fix["Courier Payments"]\
                                      - courierPayments_fix["Eater Fare Adjustments"] - courierPayments_fix["Eater Other Promotionss / Discounts"] - \
                                        courierPayments_fix["Eater Referral Promotions"]) / courierPayments_fix["Deliveries"]
courierPayments_fix["Unit Economics"] = courierPayments_fix["Unit Economics"].round(2)
courierPayments_fix["Month"] = courierPayments_fix["Date"].dt.month
courierPayments_fix
#I rest from Gross Bookings the different costs Uber eats had over these months. I did not included the Delivery fee couse the task mentioned It already was in there
```

```python
sns.set_theme(style="darkgrid")
```

```python
#Behaivor of the economical units ( losses ) over time 
sns.relplot(data = courierPayments_fix, x ="Date", y ="Unit Economics", kind="line",markers=True)
plt.xticks(rotation=45)
plt.show()

#Average losses per deliver per month
money = courierPayments_fix.groupby("Month")["Unit Economics"].mean().sort_values(ascending=False).reset_index()
money
```

```python
#Assess the pros and cons of the current trends. Where do you think these trends stem from?
#How can the city improve its profitability?
#I will assume the Restaurants Payments and Courier Payments are untouchables
x_vars = ["Deliveries","Gross Bookings","Eater Fare Adjustments","Eater Other Promotionss / Discounts", "Eater Referral Promotions"]
y_vars = ["Deliveries","Gross Bookings", "Unit Economics"]
sns.pairplot(data= courierPayments_fix, x_vars=x_vars, y_vars=y_vars, diag_kind="hist",plot_kws=dict(marker="o", linewidth=1),
    diag_kws=dict(fill=False))
plt.show()
```

```python

#Pearson correlation
courierPayments_fix_2 = courierPayments_fix[["Deliveries","Gross Bookings","Eater Fare Adjustments","Eater Other Promotionss / Discounts", "Eater Referral Promotions","Unit Economics"]]
pearson = courierPayments_fix_2.corr(method="pearson")
pearson
sns.heatmap(pearson,xticklabels=pearson.columns,
            yticklabels=pearson.columns,
            cmap="RdBu_r",
            annot=True,
            linewidths=0.5)
plt.show()

#The points fall close to the line, which indicates that there is a strong negative relationship between the variables. 
# The relationship is negative because, as one variable increases, the other variable decreases.
```

```python
#lets reduce those two columns
courierPayments_fix["Eater Other Promotionss / Discounts"] = courierPayments_fix["Eater Other Promotionss / Discounts"]/3.5
courierPayments_fix["Eater Referral Promotions"] = courierPayments_fix["Eater Referral Promotions"]/4
```

```python
#NEW UNIT ECONOMIS
courierPayments_fix["New Unit Economics"] = (courierPayments_fix["Gross Bookings"] - courierPayments_fix["Restaurant Payments"] - courierPayments_fix["Courier Payments"]\
                                      - courierPayments_fix["Eater Fare Adjustments"] - courierPayments_fix["Eater Other Promotionss / Discounts"] - \
                                        courierPayments_fix["Eater Referral Promotions"]) / courierPayments_fix["Deliveries"]

courierPayments_fix
```

```python
money_2 = courierPayments_fix.groupby("Month")["New Unit Economics"].mean().sort_values(ascending=False).reset_index()

sns.barplot(data = money_2, x="Month", y = "New Unit Economics")
plt.show()

money_2
```
